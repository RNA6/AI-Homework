```python
import random

class World:

    def vacuum(this, dirt_count):
        # grid size (10x10)
        this.n = 10  

        #Create grid filled with "c"
        this.map = [["c" for _ in range(this.n)] for _ in range(this.n)]

        # Add walls "l"
        for i in range(this.n):
            this.map[0][i] = "l"
            this.map[this.n-1][i] = "l"
            this.map[i][0] = "l"
            this.map[i][this.n-1] = "l"

        # Add obstacles "#"
        this.map[3][3] = "#"
        this.map[3][4] = "#"
        this.map[7][6] = "#"

        # Add random dirt "*"
        count = 0
        while count < dirt_count:
            r = random.randint(1, this.n-2)
            c = random.randint(1, this.n-2)
            if this.map[r][c] == "c":
                this.map[r][c] = "*"
                count += 1

        # Agent start position
        this.row = 1
        this.col = 1

    # Sensors 
    def isDirty(this):
        return this.map[this.row][this.col] == "*"

    # Actions 
    def moveRight(this):
        if this.map[this.row][this.col + 1] not in ["l", "#"]:
            this.col += 1
            return True
        return False

    def moveLeft(this):
        if this.map[this.row][this.col - 1] not in ["l", "#"]:
            this.col -= 1
            return True
        return False

    def moveUp(this):
        if this.map[this.row - 1][this.col] not in ["l", "#"]:
            this.row -= 1
            return True
        return False

    def moveDown(this):
        if this.map[this.row + 1][this.col] not in ["l", "#"]:
            this.row += 1
            return True
        return False

    def suck(this):
        if this.isDirty():
            this.map[this.row][this.col] = "c"

    def show(this):
        for r in range(this.n):
            display_row = this.map[r][:]
            if r == this.row:
                display_row[this.col] = "A"
            print(" ".join(display_row))
        


# program test
world = World()
world.vacuum(5)
world.show()
```

```python
#Common variables and functions
environment = np.full((5,5), "C")
environment[3:, 0::2] = "|"
environment[0:2, 1::2] = "|"
environment[0, 2::2] = "*"
environment[3, 1] = "*"
environment[2, 3] = "*"
environment[4, 3] = "*"

#Location
row = 0
column = 0
dirt = 5

def isDirty():
    if(environment[row][column] == "*"):
        return True
    return False

#Actions
def moveLeft():
    global row, column
    if(column > 0 and environment[row][column-1] != "|"):
        column -= 1
        return True
    return False

def moveRight():
    global row, column
    if(column < environment.shape[1] - 1 and environment[row][column+1] != "|"):
        column += 1
        return True
    return False

def moveUp():
    global row, column
    if(row > 0 and environment[row-1][column] != "|"):
        row -= 1
        return True
    return False

def moveDown():
    global row, column
    if(row < environment.shape[0] - 1 and environment[row+1][column] != "|"):
        row += 1
        return True
    return False

def suck():
    global dirt
    environment[row][column] = "C"
    dirt -= 1
print(environment)
```

```python
import random
directions = ["Left", "Right", "Up", "Down"]

print(random.choice(directions))
def simpleReflexAgent():
    visited = set() 
    while(dirt > 0):
        location = (row, column)
        if location not in visited:
            visited.add(location)
            if(isDirty()):
                suck()
            if(moveRight()):
                continue 

        canMove = False
        possibleMoves = ["Left", "Up", "Down"]     
        
        while(not canMove and len(possibleMoves) > 0):
            nextMove = random.choice(possibleMoves)
            if(nextMove == "Left"):
                if(moveLeft()):
                    canMove = True
                else:
                    possibleMoves.remove("Left")
            elif(nextMove == "Up"):
                if(moveUp()):
                    canMove = True
                else:
                    possibleMoves.remove("Up")
            elif(nextMove == "Down"):
                if(moveDown()):
                    canMove = True
                else:
                    possibleMoves.remove("Down")
            # print(possibleMoves)

print("Not cleaned Room\n" ,environment)
simpleReflexAgent()
print("Cleaned Room\n" , environment)
```
