```python
import random

class World:

    def vacuum(this, dirt_count):
        # grid size (10x10)
        this.n = 10  

        #Create grid filled with "c"
        this.map = [["c" for _ in range(this.n)] for _ in range(this.n)]

        # Add walls "l"
        for i in range(this.n):
            this.map[0][i] = "l"
            this.map[this.n-1][i] = "l"
            this.map[i][0] = "l"
            this.map[i][this.n-1] = "l"

        # Add obstacles "#"
        this.map[3][3] = "#"
        this.map[3][4] = "#"
        this.map[7][6] = "#"

        # Add random dirt "*"
        this.dirt = 0
        while this.dirt < dirt_count:
            r = random.randint(1, this.n-2)
            c = random.randint(1, this.n-2)
            if this.map[r][c] == "c":
                this.map[r][c] = "*"
                this.dirt += 1

        # Agent start position
        this.row = 1
        this.col = 1

    # Sensors 
    def isDirty(this):
        return this.map[this.row][this.col] == "*"

    # Actions 
    def moveRight(this):
        if this.map[this.row][this.col + 1] not in ["l", "#"]:
            this.col += 1
            return True
        return False

    def moveLeft(this):
        if this.map[this.row][this.col - 1] not in ["l", "#"]:
            this.col -= 1
            return True
        return False

    def moveUp(this):
        if this.map[this.row - 1][this.col] not in ["l", "#"]:
            this.row -= 1
            return True
        return False

    def moveDown(this):
        if this.map[this.row + 1][this.col] not in ["l", "#"]:
            this.row += 1
            return True
        return False

    def suck(this):
        this.map[this.row][this.col] = "c"
        this.dirt -= 1

    def show(this):
        for r in range(this.n):
            display_row = this.map[r][:]
            if r == this.row:
                display_row[this.col] = "A"
            print(" ".join(display_row))

# program test
world = World()
world.vacuum(10)
world.show()
```

```python
def randomMove():
    canMove = False
    possibleMoves = ["Left", "Up", "Down"]     
    
    while(not canMove):
        nextMove = random.choice(possibleMoves)
        if(nextMove == "Left"):
            if(world.moveLeft()):
                canMove = True
            else:
                possibleMoves.remove("Left")
        elif(nextMove == "Up"):
            if(world.moveUp()):
                canMove = True
            else:
                possibleMoves.remove("Up")
        elif(nextMove == "Down"):
            if(world.moveDown()):
                canMove = True
            else:
                possibleMoves.remove("Down")

def simpleReflexAgent():
    visited = set() 
    while(world.dirt > 0):
        location = (world.row, world.col)
        if location not in visited:
            visited.add(location)
            if(world.isDirty()):
                world.suck()
                
            if(world.moveRight()):
                continue 

print("Room before Cleaning Room")
world.show()
print()
simpleReflexAgent()
print("Room after Cleaning")
world.show()
```
